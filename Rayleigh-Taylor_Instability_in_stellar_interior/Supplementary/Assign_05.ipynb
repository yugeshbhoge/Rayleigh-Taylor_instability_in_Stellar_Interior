>### **2) & 3) Calculating number fraction and mass fraction for stars**

**Solution**

>>### **For number fraction**

Salpeter Initial Mass Function is given by,

>>>\begin{equation}
\boxed{
\frac{dN}{dm} = k \times m^{-2.35}
}
\end{equation}
where,
$k$ is proportionality constant.

By integrating it over the total mass range which is from $0.8M_{\odot}$ to $200M_{\odot}$ we will get total number of stars.
\begin{equation}
N_{Total} = \int_{0.8M_{\odot}}^{200M_{\odot}} km^{-2.35}\,dm
\tag{2}
\end{equation}
which comes to be,
$$N_{Total} = k \times \left(\frac{0.8^{-1.35}-200^{-1.35}}{1.35}\right)\times M_{\odot}^{-1.35}$$

\begin{equation}
N_{Total}\approx \frac{1.35}{1.35}M_{\odot}^{-1.35}k=kM_{\odot}^{-1.35}
\tag{3}
\end{equation}

>>>\begin{equation} 
\boxed{
N_{Total} = k \times M_{\odot}^{-1.35}
}
\tag{4}
\end{equation}

Now for finding fraction of stars in given mass ranges we will first integrate them over the given mass range and then divide it with $N_{Total}$

>>>1. For mass range $0.8M_{\odot}-8M_{\odot}$

$$N_{1} = k \times \left(\frac{0.8^{-1.35}-8^{-1.35}}{1.35}\right)\times M_{\odot}^{-1.35}$$
$$N_{1} = 0.9559 \times k \times M_{\odot}^{-1.35}$$

$$\boxed{f(N_{1}) = \frac{N_{1}}{N_{Total}} = 0.9559}$$

>>>2. For mass range $8M_{\odot}-15M_{\odot}$

$$N_{2} = k \times \left(\frac{8^{-1.35}-15^{-1.35}}{1.35}\right)\times M_{\odot}^{-1.35}$$
$$N_{2} = 0.0256 \times k \times M_{\odot}^{-1.35}$$

$$\boxed{f(N_{2}) = \frac{N_{2}}{N_{Total}} = 0.0256}$$

>>>3. For mass range $15M_{\odot}-200M_{\odot}$

$$N_{3} = k \times \left(\frac{15^{-1.35}-200^{-1.35}}{1.35}\right)\times M_{\odot}^{-1.35}$$
$$N_{3} = 0.0185 \times k \times M_{\odot}^{-1.35}$$

$$\boxed{f(N_{3}) = \frac{N_{3}}{N_{Total}} = 0.0185}$$


>>### For mass fraction
We have to multipy the the IMF by m to get the mass fraction. Hence the equation is given by,

\begin{equation}
\frac{dN_{mass}}{dm} = k \times m \times m^{-2.35}
\tag{5}
\end{equation}

>>>\begin{equation}
\boxed{
\frac{dN_{mass}}{dm} = k \times m^{-1.35}
}
\tag{6}
\end{equation}

By integrating it over the total mass range which is from $0.8M_{\odot}$ to $200M_{\odot}$ we will get total number of stars in given total mass range.
\begin{equation}
N_{mass} = \int_{0.8M_{\odot}}^{200M_{\odot}} k \times m^{-1.35}\,dm
\tag{7}
\end{equation}
which comes to be,
$$N_{mass} = k \times \left(\frac{0.8^{-0.35}-200^{-0.35}}{0.35}\right)\times M_{\odot}^{-0.35}$$

>>>\begin{equation} 
\boxed{
N_{mass} = 2.642 \times k \times M_{\odot}^{-0.35}
}
\tag{8}
\end{equation}

Now for finding mass fraction of stars in given mass ranges we will first integrate them over the given mass range and then divide it with $N_{mass}$

>>>1. For mass range $0.8M_{\odot}-8M_{\odot}$

$$N_{1}^{mass} = k \times \left(\frac{0.8^{-0.35}-8^{-0.35}}{0.35}\right)\times M_{\odot}^{-0.35}$$
$$N_{1}^{mass} = 1.709 \times k \times M_{\odot}^{-0.35}$$

$$\boxed{f(N_{1}^{mass}) = \frac{N_{1}^{mass}}{N_{mass}} = 0.647}$$

>>>2. For mass range $8M_{\odot}-15M_{\odot}$

$$N_{2}^{mass} = k \times \left(\frac{8^{-0.35}-15^{-0.35}}{0.35}\right)\times M_{\odot}^{-0.35}$$
$$N_{2}^{mass} = 0.2726 \times k \times M_{\odot}^{-0.35}$$

$$\boxed{f(N_{2}^{mass}) = \frac{N_{2}^{mass}}{N_{mass}} = 0.103}$$

>>>3. For mass range $15M_{\odot}-200M_{\odot}$

$$N_{3}^{mass} = k \times \left(\frac{15^{-0.35}-200^{-0.35}}{0.35}\right)\times M_{\odot}^{-0.35}$$
$$N_{3}^{mass} = 0.6601 \times k \times M_{\odot}^{-0.35}$$

$$\boxed{f(N_{3}^{mass}) = \frac{N_{3}^{mass}}{N_{mass}} = 0.250}$$

```python
import math
from scipy.integrate import quad
import matplotlib.pyplot as plt
#Exponent
ex = 2.35
#IMF
def imf(m):
    return m**(-ex)
#IMF mass
def imf_m(m):
    return m * m**(-ex)

m_min = 0.8
m_max = 200.0
#mass ranges
m_bins = {
    "0.8-8": (0.8, 8),
    "8-15": (8, 15),
    "15-200": (15, 200)
}
#
total_num, err_total = quad(imf, m_min, m_max)
total_mass, err_mass = quad(imf_m, m_min, m_max)

print("Total number (normalized):", total_num)
print("Total mass (normalized):", total_mass)
print("")
num_frac=[]
mass_frac=[]
# Compute the number and mass fractions for each mass bin
for label, (m_low, m_high) in m_bins.items():
    # Integrate number in this bin
    num_bin, err_bin = quad(imf, m_low, m_high)
    # Integrate mass in this bin
    mass_bin, err_mass_bin = quad(imf_m, m_low, m_high)
    
    n_frac = num_bin / total_num
    m_frac = mass_bin / total_mass
    # Append to lists
    num_frac.append(n_frac)
    mass_frac.append(m_frac)

    print(f"Mass Range {label} M\u2609:")
    print(f"  Number Fraction = {n_frac:.4f}")
    print(f"  Mass Fraction   = {m_frac:.4f}")
    print("")
plt.figure(figsize=(10, 6))
plt.plot(m_bins.keys(), num_frac, label='Number Fraction', marker='o')
plt.plot(m_bins.keys(), mass_frac, label='Mass Fraction', marker='o')
plt.xlabel('Mass Range (M\u2609)')
plt.ylabel('Fraction')
plt.title('Number and Mass Fractions (using Salpeter IMF)')
plt.legend()
plt.grid()
# plt.savefig('imf_fractions.png')
plt.show()
```
