## Q1
Starting from
$$
J = \left| \frac{\partial x^i}{\partial \xi^j} \right| = \epsilon^{ijk} \frac{\partial x^{(1)}}{\partial \xi^i} \frac{\partial x^{(2)}}{\partial \xi^j} \frac{\partial x^{(3)}}{\partial \xi^k},
$$
and commuting relation that
$$
\frac{D}{Dt} \left( \frac{\partial x^i}{\partial \xi^j} \right) = \frac{\partial}{\partial \xi^j} \frac{D x^i}{D t} = \frac{\partial v^i}{\partial \xi^j},
$$
prove the Euler's expansion formula --
$$
\frac{DJ}{Dt} = \left( \frac{\partial v^{(1)}}{\partial x^{(1)}} + \frac{\partial v^{(2)}}{\partial x^{(2)}} + \frac{\partial v^{(3)}}{\partial x^{(3)}} \right) J = (\nabla \cdot \mathbf{v}) J, \quad \text{Euler’s expansion formula.}
$$

## Solution:


### 1. Definitions

  $$
    \mathbf{x}(t,\boldsymbol{\xi}) 
    \;=\; 
    \bigl(x^{(1)}(t,\boldsymbol{\xi}),\,x^{(2)}(t,\boldsymbol{\xi}),\,x^{(3)}(t,\boldsymbol{\xi})\bigr).
  $$

  The **Jacobian** of this mapping with respect to the Lagrangian coordinates $\boldsymbol{\xi} = (\xi^1,\xi^2,\xi^3)$ is  

  $$
    J
    \;=\;
    \left|\frac{\partial \mathbf{x}}{\partial \boldsymbol{\xi}}\right|
    \;=\;
    \epsilon^{i j k}\,\frac{\partial x^{(1)}}{\partial \xi^i}\,\frac{\partial x^{(2)}}{\partial \xi^j}\,\frac{\partial x^{(3)}}{\partial \xi^k}.
  $$

- **Velocity**:  

  $$
    \frac{d}{dt}\,\mathbf{x}(t,\boldsymbol{\xi}) 
    \;=\; 
    \mathbf{v}\bigl(t,\,\mathbf{x}(t,\boldsymbol{\xi})\bigr).
  $$

- **Total derivative**: for a scalar function $F(t,\mathbf{x})$,  

  $$
    \frac{D F}{D t} \;=\; \frac{\partial F}{\partial t} \;+\; (\mathbf{v}\cdot\nabla)F.
  $$

- **Commutataion relation**:  

  $$
    \frac{D}{D t}\!\Bigl(\tfrac{\partial x^i}{\partial \xi^j}\Bigr)
    \;=\;
    \frac{\partial}{\partial \xi^j}\!\Bigl(\tfrac{D x^i}{D t}\Bigr)
    \;=\;
    \frac{\partial v^i}{\partial \xi^j}.
  $$


### 2. Differentiating $J$ in Time  


$$
  \frac{D J}{D t}
  \;=\;
  \frac{D}{D t}
  \Bigl(\,
    \epsilon^{i j k}
    \,\frac{\partial x^{(1)}}{\partial \xi^i}
    \,\frac{\partial x^{(2)}}{\partial \xi^j}
    \,\frac{\partial x^{(3)}}{\partial \xi^k}
  \Bigr).
$$

As $\epsilon^{i j k}$ is constant,  

$$
  \frac{D J}{D t}
  \;=\;
  \epsilon^{i j k}
  \Bigl[
    \frac{D}{D t}\Bigl(\tfrac{\partial x^{(1)}}{\partial \xi^i}\Bigr)\,\tfrac{\partial x^{(2)}}{\partial \xi^j}\,\tfrac{\partial x^{(3)}}{\partial \xi^k}
    \;+\;
    \tfrac{\partial x^{(1)}}{\partial \xi^i}\,\frac{D}{D t}\Bigl(\tfrac{\partial x^{(2)}}{\partial \xi^j}\Bigr)\,\tfrac{\partial x^{(3)}}{\partial \xi^k}
    \;+\;
    \tfrac{\partial x^{(1)}}{\partial \xi^i}\,\tfrac{\partial x^{(2)}}{\partial \xi^j}\,\frac{D}{D t}\Bigl(\tfrac{\partial x^{(3)}}{\partial \xi^k}\Bigr)
  \Bigr].
$$

By the commuting relation,  

$$
  \frac{D}{D t}\!\Bigl(\tfrac{\partial x^p}{\partial \xi^q}\Bigr)
  \;=\;
  \frac{\partial v^p}{\partial \xi^q},
$$

putting above we get,  

$$
  \frac{D J}{D t}
  \;=\;
  \epsilon^{i j k}
  \Bigl[
    \tfrac{\partial v^{(1)}}{\partial \xi^i}\,\tfrac{\partial x^{(2)}}{\partial \xi^j}\,\tfrac{\partial x^{(3)}}{\partial \xi^k}
    \;+\;
    \tfrac{\partial x^{(1)}}{\partial \xi^i}\,\tfrac{\partial v^{(2)}}{\partial \xi^j}\,\tfrac{\partial x^{(3)}}{\partial \xi^k}
    \;+\;
    \tfrac{\partial x^{(1)}}{\partial \xi^i}\,\tfrac{\partial x^{(2)}}{\partial \xi^j}\,\tfrac{\partial v^{(3)}}{\partial \xi^k}
  \Bigr].
$$

### 3. Converting Derivatives w.r.t. $\xi$ to Derivatives w.r.t. $x$  

Also, 

$$
  \frac{\partial}{\partial \xi^j}
  \;=\;
  \frac{\partial x^m}{\partial \xi^j}
  \,\frac{\partial}{\partial x^m}
  \quad\Longrightarrow\quad
  \frac{\partial v^{(p)}}{\partial \xi^j}
  \;=\;
  \frac{\partial v^{(p)}}{\partial x^m}
  \,\frac{\partial x^m}{\partial \xi^j}.
$$

Putting this we get, 

$$
  \frac{D J}{D t}
  \;=\;
  J \Bigl(
  \tfrac{\partial v^{(1)}}{\partial x^{(1)}}
  + \tfrac{\partial v^{(2)}}{\partial x^{(2)}}
  + \tfrac{\partial v^{(3)}}{\partial x^{(3)}}
  \Bigr).
$$

Also,  $\tfrac{\partial v^{(1)}}{\partial x^{(1)}} + \tfrac{\partial v^{(2)}}{\partial x^{(2)}} + \tfrac{\partial v^{(3)}}{\partial x^{(3)}} = \nabla \cdot \mathbf{v}$,

$$
  \boxed{
    \frac{D J}{D t}
    \;=\;
    (\nabla \cdot \mathbf{v})\,J.
  }
$$

## Q6 
Consider a steady axisymmetric flow of a compressible fluid. The equation of continuity in cylindrical coordinates $(R, \varphi, x)$  is

$$
\frac{\partial}{\partial R} (\rho R u_R) + \frac{\partial}{\partial x} (\rho R u_x) = 0
$$

Show how we can define a streamfunction so that the equation of continuity is satisfied automatically.

Using the indicial notation (and without using any vector identity) show that the acceleration of a fluid particle is given by,

$$
\mathbf{a} = \frac{\partial \mathbf{u}}{\partial t} + \nabla \left(\frac{1}{2} q^2\right) + \boldsymbol{\omega} \times \mathbf{u}
$$

where  q  is the magnitude of velocity  $\mathbf{u}$  and  $\boldsymbol{\omega} = \nabla \times \mathbf{u}$  is the vorticity.

## Solution


### 1. Continuity \& streamfunction in axisymmetric flow

\begin{equation}
\frac{\partial}{\partial R}\bigl(\rho R\,u_{R}\bigr)
\;+\;
\frac{\partial}{\partial x}\bigl(\rho R\,u_{x}\bigr)
\;=\;0
\end{equation}

Define a streamfunction $\Psi(R,x)$ such that,
\begin{equation}
\rho R \, u_{R} \;=\; \frac{\partial \Psi}{\partial x},
\quad
\rho R \, u_{x} \;=\; -\,\frac{\partial \Psi}{\partial R}.
\end{equation}

Then the above partial derivatives sum to zero automatically.

### 2. Acceleration in index notation

Let $u_{i}$  be velocity components and  $a_{i}$  the acceleration. Then:

\begin{equation}
a_{i}
\;=\;
\frac{D u_{i}}{D t}
\;=\;
\frac{\partial u_{i}}{\partial t}
\;+\;
u_{j}\,\frac{\partial u_{i}}{\partial x_{j}}.
\end{equation}

Write  $u_{j}\,\partial u_{i}/\partial x_{j}$  as:

\begin{equation}
u_{j}\,\frac{\partial u_{i}}{\partial x_{j}}
\;=\;
\frac{\partial}{\partial x_{i}}\!\biggl(\tfrac{1}{2}\,q^{2}\biggr)
\;+\;
\epsilon_{i j k}\,\omega_{j}\,u_{k},
\quad
\text{where }
q^{2} = u_{m}\,u_{m},
\;
\omega_{j} = \epsilon_{j l m}\,\frac{\partial u_{m}}{\partial x_{l}}.
\end{equation}

Hence:

\begin{equation}
a_{i}
\;=\;
\frac{\partial u_{i}}{\partial t}
\;+\;
\frac{\partial}{\partial x_{i}}
\bigl(\tfrac{1}{2}\,q^{2}\bigr)
\;+\;
\epsilon_{i j k}\,\omega_{j}\,u_{k}.
\end{equation}

In vector notion:

\begin{equation}
\mathbf{a}
\;=\;
\frac{\partial \mathbf{u}}{\partial t}
\;+\;
\nabla\Bigl(\tfrac{1}{2}\,q^{2}\Bigr)
\;+\;
\boldsymbol{\omega}\times\mathbf{u}.
\end{equation}

## Q7
Solve the expression where curl operator is applied on both sides of Navier-Stokes Equation to write the equivalent equation in terms of vorticity, velocity and viscosity.  $\phi$  is gravitational potential.

$$
\nabla \times \left\{ 
\frac{\partial \mathbf{u}}{\partial t} 
+ \mathbf{u} \cdot \nabla \mathbf{u} 
= -\frac{1}{\rho} \nabla p 
+ \nabla \phi 
+ \nu \nabla^2 \mathbf{u} 
\right\}
$$


## Solution:

Starting Navier-Stokes:
\begin{equation}
    \frac{\partial \mathbf{u}}{\partial t}
    \;+\;
    (\mathbf{u}\cdot\nabla)\mathbf{u}
    \;=\;
    -\frac{1}{\rho}\,\nabla p
    \;+\;
    \nabla\phi
    \;+\;
    \nu\,\nabla^{2}\mathbf{u}.
\end{equation}

Apply  $\nabla\times$  to both sides:
\begin{equation}
    \nabla \times \left(\frac{\partial \mathbf{u}}{\partial t}
    \;+\;
    (\mathbf{u}\cdot\nabla)\mathbf{u}\right)
    \;=\;
    \nabla \times \left(-\frac{1}{\rho}\,\nabla p
    \;+\;
    \nabla\phi
    \;+\;
    \nu\,\nabla^{2}\mathbf{u}\right).
\end{equation}

Note:  $\nabla \times (\nabla\ \cdot)=0$  for any scalar field, so the  p  and  $\phi$  terms vanish. Also assume constant  $\nu$ :
\begin{equation}
    \frac{\partial}{\partial t} \bigl(\nabla\times\mathbf{u}\bigr)
    \;+\;
    \nabla \times \bigl[(\mathbf{u}\cdot\nabla)\mathbf{u}\bigr]
    \;=\;
    \nu\,\nabla \times \nabla^{2}\mathbf{u}.
\end{equation}

Define vorticity  $\boldsymbol{\omega} = \nabla\times\mathbf{u}$. Then,
\begin{equation}
    \nabla \times \nabla^{2}\mathbf{u}
    \;=\;
    \nabla^{2}(\nabla\times\mathbf{u})
    \;=\;
    \nabla^{2}\boldsymbol{\omega}.
\end{equation}

\begin{equation}
    \frac{\partial \boldsymbol{\omega}}{\partial t}
    \;+\;
    \nabla \times \bigl[(\mathbf{u}\cdot\nabla)\mathbf{u}\bigr]
    \;=\;
    \nu\,\nabla^{2}\boldsymbol{\omega}.
\end{equation}

Use the identity
\begin{equation}
(\mathbf{u}\cdot\nabla)\mathbf{u}
\;=\;
-\,\mathbf{u}\times(\nabla\times\mathbf{u})
\;+\;
\tfrac{1}{2}\nabla|\mathbf{u}|^{2},
\end{equation}
and
$\nabla \times \bigl(\nabla\,\frac{|\mathbf{u}|^{2}\!}{2}\bigr)=0$.
Hence
\begin{equation}
    \nabla \times \bigl[(\mathbf{u}\cdot\nabla)\mathbf{u}\bigr]
    \;=\;
    -\,\nabla \times \bigl[\mathbf{u}\times\boldsymbol{\omega}\bigr].
\end{equation}

Expanding,
\begin{equation}
    \nabla \times \bigl(\mathbf{u}\times\boldsymbol{\omega}\bigr)
    =
    \mathbf{u}\,(\nabla\cdot\boldsymbol{\omega})
    -\boldsymbol{\omega}\,(\nabla\cdot\mathbf{u})
    +(\boldsymbol{\omega}\cdot\nabla)\mathbf{u}
    -(\mathbf{u}\cdot\nabla)\boldsymbol{\omega}.
\end{equation}

Vorticity equation:
\begin{equation}
    \frac{\partial \boldsymbol{\omega}}{\partial t}
    \;+\;
    (\mathbf{u}\cdot\nabla)\boldsymbol{\omega}
    \;=\;
    (\boldsymbol{\omega}\cdot\nabla)\mathbf{u}
    \;-\;
    \boldsymbol{\omega}\,(\nabla\cdot\mathbf{u})
    \;+\;
    \nu\,\nabla^{2}\boldsymbol{\omega}.
\end{equation}

If incompressible ($\nabla\cdot\mathbf{u}=0$):

\begin{equation}
\boxed{
    \frac{\partial \boldsymbol{\omega}}{\partial t}
    \;+\;
    (\mathbf{u}\cdot\nabla)\boldsymbol{\omega}
    \;=\;
    (\boldsymbol{\omega}\cdot\nabla)\mathbf{u}
    \;+\;
    \nu\,\nabla^{2}\boldsymbol{\omega}.
}
\end{equation}

## Q9 
If Sun is losing mass, what could be the rate of it by measuring the the mass density and velocity of interplanetary medium just outside Earth. You start from the continuity equation and assume spherically symmetric flow.

## Solution:

\begin{equation}
    \frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{v}) = 0
\end{equation}

In steady state,

\begin{equation}
    \frac{\partial \rho}{\partial t} = 0 \implies \nabla \cdot (\rho \mathbf{v}) = 0
\end{equation}

For spherical symmetry, $\rho=\rho(r)$ and $\mathbf{v} = v(r)\hat{r}$. Hence,

\begin{equation}
    \nabla \cdot (\rho \mathbf{v}) 
    = \frac{1}{r^2}\frac{\partial}{\partial r}(r^2 \rho(r) v(r)) = 0
\end{equation}

which implies

\begin{equation}
    \rho(r) \, v(r) \, 4\pi r^2 = \text{constant} = \frac{d M}{d t}
\end{equation}

Therefore, by measuring $\rho$ and $v$ at Earth's orbit $r$, the mass loss rate of the Sun is

\begin{equation}
    \frac{d M}{d t} = 4\pi \, r^2 \, \rho(r) \, v(r)
\end{equation}

## Q 10
Earth's atmosphere Use the continuity equation and ideal gas law for steady state case to find the pressure variation with respect to height from Earth's surface. Assume constant temperature vs altitude.

## Solution
General continuity (steady state flow):
\begin{equation}
 \frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{v}) = 0 \implies \nabla \cdot (\rho \mathbf{v}) = 0
\end{equation}

\begin{equation}
\text{Static fluid: } \mathbf{v} = 0 \implies \nabla \cdot 0 = 0
\end{equation}

\begin{equation}
\text{Momentum (no motion): } 0 = -\nabla p + \rho \mathbf{g} \implies \frac{dp}{dz} = -\rho g
\end{equation}

\begin{equation}
\text{Ideal gas law: } \rho = \frac{p M}{R T} \quad (\text{constant }T)
\end{equation}

\begin{equation}
\frac{dp}{dz} = -\frac{M g}{R T} \, p \implies \frac{dp}{p} = - \frac{M g}{R T} \, dz

## Q11

## Solution:


\begin{equation}
    \text{Continuity equation: }  \frac{\partial \rho}{\partial t} + \nabla \cdot \bigl(\rho \mathbf{v}\bigr) = 0
\end{equation}

\text{Steady: } \frac{\partial \rho}{\partial t}=0 \implies \nabla \cdot \bigl(\rho \mathbf{v}\bigr)=0
\end{equation}

\end{equation}

## Q12
Write down the hydrostatic stellar structure equation in spherical coordinate system where the acceleration due to gravity if expressed in the form gravitational potential using Poisson's equation,
$$
\nabla^2 \phi = 4\pi G \rho
$$

Use polytropic equation of state,

$$
PV^n = \text{const.} \equiv P = K \rho^\gamma
$$

to arrive at the dimensionless \textit{Lane-Emden Equation} using dimensionless quantities,  $\theta$  and  $\xi$. These are defined as  $\rho = \rho_c \theta^n$  and  $r = \alpha \xi$  and  $\gamma = 1 + \frac{1}{n}$

$$
\frac{1}{\xi^2} \frac{d}{d\xi} \left( \xi^2 \frac{d\theta}{d\xi} \right) = -\theta^n
$$


## Solution

(1) Hydrostatic Equilibrium in Spherical Symmetry

\begin{equation}
\frac{dP}{dr} = -\,\rho \,\frac{G\,M(r)}{r^2},
\end{equation}

where 
$M(r) = 4\pi \displaystyle \int_{0}^{r} \rho(r')\,r'^{2}\,dr'$.

(2) Polytropic Equation of State

\begin{equation}
P = K\,\rho^\gamma, 
\quad
\gamma = 1 + \frac{1}{n}.
\end{equation}

Hence 

$\rho^{\gamma-1} = \rho^{\tfrac{1}{n}}$. 
Also, from $P = K\,\rho^\gamma$,

\begin{equation}
\frac{dP}{dr} 
= 
K\,\gamma\,\rho^{\gamma-1}\,\frac{d\rho}{dr}.
\end{equation}

(3) Combine Hydrostatic Equilibrium with Polytropic Relation

Equating the two expressions for $\displaystyle \frac{dP}{dr}$,

\begin{equation}
K\,\gamma\,\rho^{\gamma-1}\,\frac{d\rho}{dr}
= 
-\,\rho\,\frac{G\,M(r)}{r^2}.
\tag{1}
\end{equation}

(4) Define Dimensionless Variables

Set 
\begin{equation}
\rho = \rho_c\,\theta^n,
\quad
n = \frac{1}{\gamma - 1},
\quad
r = \alpha\,\xi,
\end{equation}

where 
\begin{equation}
\alpha^2 = \frac{K\,(n+1)\,\rho_c^{\tfrac{1}{n}-1}}{4\pi\,G},
\quad
\rho_c = \rho(r=0).
\end{equation}

Then

\begin{equation}
\frac{d\rho}{dr} 
= 
\rho_c\,n\,\theta^{\,n-1}\,\frac{d\theta}{dr} 
\quad\text{and}\quad
\rho^{\,\gamma-1} = (\rho_c\,\theta^n)^{\tfrac{1}{n}} = \rho_c^{\tfrac{1}{n}}\,\theta.
\end{equation}

(5) Substitute into (1)

Left side of (1):
$$
K\,\gamma\,\rho^{\gamma-1}\,\frac{d\rho}{dr}
=
K\,\gamma\,\rho_c^{\tfrac{1}{n}}\,\theta
\,\Bigl[\rho_c\,n\,\theta^{\,n-1}\,\frac{d\theta}{dr}\Bigr]
=
K\,\gamma\,\rho_c^{\tfrac{1}{n}+1}\,n\,\theta^n\,\frac{d\theta}{dr}.
$$

Right side of (1):
$$
-\rho\,\frac{G\,M(r)}{r^2}
=
-\rho_c\,\theta^n \,\frac{G\,M(r)}{r^2}.
$$

Hence

\begin{equation}
K\,\gamma\,\rho_c^{\tfrac{1}{n}+1}\,n\,\theta^n\,\frac{d\theta}{dr}
=
-\,\rho_c\,\theta^n\,\frac{G\,M(r)}{r^2}.
\end{equation}

Divide by $\rho_c\,\theta^n$:

\begin{equation}
K\,\gamma\,\rho_c^{\tfrac{1}{n}}\,n\,\frac{d\theta}{dr}
=
-\,\frac{G\,M(r)}{r^2}.
\tag{2}
\end{equation}

(6) Express $M(r)$ and $r^2$ in Terms of $\xi$

\begin{equation}
M(r) 
= 
4\pi \int_{0}^{r}\!\rho(r')\,r'^{\,2}\,dr'
= 
4\pi\,\rho_c\,\alpha^3 \int_{0}^{\xi}\!\theta^n(\xi')\,\xi'^{\,2}\,d\xi',
\quad
r^2 = \alpha^2\,\xi^2.
\end{equation}

So

\begin{equation}
\frac{M(r)}{r^2}
=
\frac{
4\pi\,\rho_c\,\alpha^3 \displaystyle \int_{0}^{\xi}\!\theta^n(\xi')\,\xi'^{\,2}\,d\xi'
}
{\alpha^2\,\xi^2}
=
4\pi\,\rho_c\,\alpha\,\frac{1}{\xi^2}
\int_{0}^{\xi}\!\theta^n(\xi')\,\xi'^{\,2}\,d\xi'.
\end{equation}

(7) Rewrite (2) in Terms of $\xi$ and $\theta$

Since $\displaystyle \frac{d\theta}{dr} = \frac{1}{\alpha}\,\frac{d\theta}{d\xi}$,

\begin{equation}
K\,\gamma\,\rho_c^{\tfrac{1}{n}}\,n
\left(\frac{1}{\alpha}\,\frac{d\theta}{d\xi}\right)
=
-\,G
\left[
4\pi\,\rho_c\,\alpha
\frac{1}{\xi^2}
\int_{0}^{\xi}\!\theta^n(\xi')\,\xi'^{2}\,d\xi'
\right].
\end{equation}

Factor out constants. Note that:

\begin{equation}
K\,(n+1)\,\rho_c^{\tfrac{1}{n}} = 4\pi\,G\,\rho_c\,\alpha^2,
\quad
\gamma = 1 + \frac{1}{n},
\quad
\gamma\,n = n + 1.
\end{equation}

Hence $K\,\gamma\,\rho_c^{\tfrac{1}{n}}\,n = K\,(n+1)\,\rho_c^{\tfrac{1}{n}}$. 
Also $4\pi\,G\,\rho_c\,\alpha^2 = K\,(n+1)\,\rho_c^{\tfrac{1}{n}}$. 

They cancel, giving us

\begin{equation}
\frac{d\theta}{d\xi}
=
-\;\frac{1}{\xi^2}
\int_{0}^{\xi}\!\theta^n(\xi')\,\xi'^{\,2}\,d\xi'.
\tag{3}
\end{equation}

Differentiate (3) w.r.t. $\xi$:

\begin{equation}
\frac{d^2\theta}{d\xi^2}
=
-\,\frac{d}{d\xi}
\left[
\frac{1}{\xi^2}
\int_{0}^{\xi}\!\theta^n(\xi')\,\xi'^{\,2}\,d\xi'
\right].
\end{equation}

Use product rule:

\begin{equation}
\frac{d}{d\xi}\!\Bigl(\tfrac{1}{\xi^2}\Bigr)
=
-\,2\,\frac{1}{\xi^3},
\quad
\text{and}
\quad
\frac{d}{d\xi}\!\Bigl(\!\int_{0}^{\xi} f(\xi')\,d\xi'\Bigr)
=
f(\xi).
\end{equation}

Let $f(\xi') = \theta^n(\xi')\,\xi'^{\,2}$. Then


\begin{equation}
\frac{d^2\theta}{d\xi^2}
=
-\,\left[
-2\,\frac{1}{\xi^3}\!\int_{0}^{\xi} f(\xi')\,d\xi'
\;+\;
\frac{1}{\xi^2}\,f(\xi)
\right]
=
2\,\frac{1}{\xi^3}
\int_{0}^{\xi}\!f(\xi')\,d\xi'
\;-\;
\frac{f(\xi)}{\xi^2}.
\end{equation}

But $f(\xi) = \theta^n(\xi)\,\xi^2$, and from (3),
$$
\int_{0}^{\xi} f(\xi')\,d\xi' 
= 
-\,\xi^2\,\frac{d\theta}{d\xi}.
$$

Therefore,

\begin{equation}
\frac{d^2\theta}{d\xi^2}
=
2\,\frac{1}{\xi^3}\Bigl[-\,\xi^2\,\frac{d\theta}{d\xi}\Bigr]
-\frac{\theta^n(\xi)\,\xi^2}{\xi^2}
=
-2\,\frac{1}{\xi}\,\frac{d\theta}{d\xi}
-\,\theta^n(\xi).
\end{equation}

Rewrite:

\begin{equation}
\frac{1}{\xi^2} \frac{d}{d\xi}\!\Bigl(\xi^2\,\frac{d\theta}{d\xi}\Bigr)
=
-\,\theta^n.
\end{equation}

$$
\boxed{
\text{Lane-Emden Equation: }
\quad
\frac{1}{\xi^2}\,\frac{d}{d\xi}
\bigl(
\xi^2 \,\frac{d\theta}{d\xi}
\bigr)
=
-\,\theta^n.
}
$$

## Q13
Show that n cannot be arbitrary for fine size stars. What is the correct range of possible values of n.

## Q14
Solve Lane-Emden equation numerically for n = 2, 3, 4, 4.999. Draw their mass-radius diagrams by choosing a range of central densities.

## Solution:

Lane-Emden equation and boundary condition,
\begin{equation}
\frac{1}{\xi^2}\,\frac{d}{d\xi}\!\Bigl(\xi^2\,\frac{d\theta}{d\xi}\Bigr)
=-\,\theta^{\,n}
\end{equation}
At the center of the star ($\xi=0$), the density $\rho$ is maximum. Hence $\theta(0)=1,\;\theta'(0)=0.$

Behavior at large  $\xi$ (asymptotics).
From the graph we can see that for n = 5,  $\theta \to 0$ but never becomes $0$ i.e we don't get any value for $\xi$ .

For  $n<5,\text{ the solution } \theta(\xi) \text{ vanishes at finite } \xi \text{ (finite radius).}$
For  $n\ge 5,\text{ no finite-radius zero; star extends to infinite size.}$

$\textbf{Hence valid range: } 0\le n<5$.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def f(x, y, n):
    t, dt = y
    if x == 0: return [dt, 0]
    return [dt, -2.*dt/x - t**n]

def e(x, y, n):
    return y[0]
e.terminal = True
e.direction = -1

def slv(n, xM=20, ms=0.01):
    y0 = [1., 0.]
    s = solve_ivp(f, [1e-8, xM], y0, args=(n,), events=e, dense_output=True, max_step=ms)
    if s.t_events[0].size > 0:
        x1 = s.t_events[0][0]
    else:
        x1 = s.t[-1]
    xg = np.linspace(0, x1, 2000)
    th = s.sol(xg)[0]
    m = np.trapz(xg**2 * (th**n), xg)
    return xg, th, x1, m, s


def mr(n, x1, m, rc, K=1., G=1.):
    cM = 4*np.pi*m*((n+1)*K/(4*np.pi*G))**1.5
    cR = x1*np.sqrt((n+1)*K/(4*np.pi*G))
    Mv, Rv = [], []
    for r in rc:
        Mv.append(cM*(r**(1+1.5/n)))
        Rv.append(cR*(r**(0.5/n)))
    return np.array(Mv), np.array(Rv)

nL = [2, 3, 4, 4.999]
rC = np.logspace(-2, 2, 100)
nL1 = [0, 1, 2, 3, 4, 5, 6]
plt.figure()
for n in nL1:
    xg, th, x1, m, s = slv(n)
    plt.plot(xg, th, label=f'n={n}')
    if s.t_events[0].size > 0: 
        plt.plot(s.t_events[0], s.sol(s.t_events[0])[0], 'ko')
plt.legend()
plt.xlabel('x')
plt.ylabel('theta')
plt.show()

plt.figure()
for n in nL:
    xg, th, x1, m, s = slv(n)
    Mv, Rv = mr(n, x1, m, rC)
    plt.plot(Rv, Mv, label=f'n={n}')
plt.xlabel('R')
plt.ylabel('M')
plt.legend()
plt.show()
```
