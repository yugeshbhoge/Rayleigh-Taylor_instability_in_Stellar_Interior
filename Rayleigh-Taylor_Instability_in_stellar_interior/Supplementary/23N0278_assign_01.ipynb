## Q1) Find few other classical methods to measure the radius of the Earth from differnet civilizations in different space and time? Summarize their relative feasability and accuracy.
---
## 1. Eratosthenes’ Shadow Method (3rd Century BCE)

### **Idea (in simple terms)**
Eratosthenes noticed that on the summer solstice, the Sun was directly overhead in Syene (no shadow at noon), but in Alexandria (roughly 800 km north), the Sun’s rays had a slight angle and caused a shadow.

### **Key Formula**
If $\Delta \theta$ is the measured angle between the Sun’s rays at the two locations, and $D$ is the distance between them, then Earth’s *circumference* $C$ is,
$$ C = \frac{360^\circ}{\Delta \theta} \times D $$

Because $C = 2\pi R$ where $R$ is Earth’s radius, we have
$$ R = \frac{C}{2\pi} $$

### **Example Calculation**
- Angle $\Delta \theta \approx 7.2^\circ$.
- Distance $D \approx 800\ \text{km}$.

Then
$$ C = \frac{360^\circ}{7.2^\circ} \times 800\ \text{km} 
  = 50 \times 800\ \text{km} 
  = 40{,}000\ \text{km}.
$$
Hence,
$$
R = \frac{40{,}000\ \text{km}}{2\pi} 
  \approx 6366\ \text{km}.
$$

### **Accuracy**
Depending on the exact distance unit (the ancient “stadion”), Eratosthenes’ result could be within about 1–2% of the modern value (Earth’s mean radius $\approx 6378\ \text{km}$). Other estimates say 1–15% because of uncertainty in the *stadion* length.


## 2. Al-Ma’mun’s Arc Measurement (9th Century CE)

### **Idea**
Astronomers under Caliph Al-Ma’mun measured the distance for a 1-degree change in latitude $d$ on a flat plain. Once they knew $d$ for $1^\circ$, they multiplied by 360 to get the full circumference.

### **Key Formulas**
$$
\text{If } d \text{ is the distance for } 1^\circ,\quad
C = 360 \times d,\quad
R = \frac{C}{2\pi}.
$$

Or, since $1^\circ = \frac{\pi}{180}$ radians, from the spherical relationship
$$
\Delta \phi = \frac{d}{R},
$$
we get
$$
R = \frac{d}{\Delta \phi} 
  = \frac{d}{\pi/180} 
  = \frac{180}{\pi} \, d.
$$

### **Accuracy**
They likely got within a few percent of the true Earth size, which is impressive for 9th-century instruments.


## 3. Al-Biruni’s Mountain Horizon Method (11th Century CE)

### **Idea**
Al-Biruni used a mountain’s height $h$ and measured the *dip to the horizon* angle $\alpha$. Then he applied right-triangle geometry relating Earth’s center, the horizon point, and the observer on the mountain.

### **Key Formula**
$$
\cos(\alpha) = \frac{R}{R + h}
\quad\Longrightarrow\quad
R = \frac{h}{1 - \cos(\alpha)}.
$$

### **Example Calculation**
- Suppose $h = 1{,}000\ \text{m}$, $\alpha = 1^\circ$.
- Then $\cos(1^\circ) \approx 0.99985$, so
  $$
  1 - \cos(1^\circ) \approx 0.00015.
  $$
  $$
  R \approx \frac{1{,}000\ \text{m}}{0.00015} 
    = 6{,}666{,}666\ \text{m} 
    \approx 6667\ \text{km}.
  $$
This is roughly 4.5% off from 6378 km. With careful measurement, Al-Biruni reported getting closer to 1–3% accuracy.


## 4. Chinese Astronomical and Survey Methods (Song Dynasty, 11th Century CE)

### **Idea**
Chinese scholars used similar shadow and latitude methods (gnomons, advanced astronomy). Though detailed records of a single, large Earth-radius measurement are sparse, the tools *could* yield results within a few percent if done carefully.

### **Formula**
Same principle as Eratosthenes or Al-Ma’mun:
$$
C = \frac{360^\circ}{\Delta \theta} \times D
\quad \text{or} \quad
C = 360 \times d_{\text{(1 deg)}}.
$$

### **Accuracy**
Exact numbers are not well documented; likely in the same “few percent” ballpark if systematically done.


## 5. Jean-Félix Picard’s Survey (17th Century CE)

### **Idea**
Picard used more advanced tools (telescopes, accurate angles) and triangulation on a meridian arc in France. He found the distance per degree of latitude, then scaled up.

### **Accuracy**
Picard’s work was within about 0.5–1% of the modern value, a major leap in precision and a bridge to modern geodesy.

## Comparative Table of Accuracy (Approx.)

| **Method**                  | **Approx. Accuracy vs. Modern Radius**     |

| Song Dynasty (11th CE)     | Not precisely recorded, likely a few percent |

## Q2) Summarize the development of accuracy of 1 AU over last two millenia.
---
## 1. Ancient Greek & Medieval Estimates

### **Idea**
- Early astronomers, like Aristarchus (3rd century BCE) and Ptolemy (2nd century CE), tried to estimate the Sun–Earth distance using geometry and observations of the Moon’s phases or eclipses.
- Their methods relied on angles and ratios between the Moon and the Sun as seen from Earth.

### **Approximate Results**
- Aristarchus suggested the Sun was around 18–20 times farther away than the Moon (the real factor is about 400).
- Later refinements in medieval times still had large uncertainties.

### **Accuracy**
- In modern terms, these ancient results could be off by factors of 5 to 10 (or more).
- Still, they established the concept that the Sun was very far away—much farther than the Moon.


## 2. 17th Century: Early Parallax Measurements

### **Idea**
- With the rise of telescopes in the early 1600s, astronomers used $parallax$ (the shift in a planet’s apparent position when viewed from different points on Earth) to estimate distances.
- Giovanni Cassini (late 17th century) measured the $parallax$ of Mars and used geometry to infer 1 AU (the Earth–Sun distance).

### **Key Concept**
$$
\text{Angular shift} \approx \frac{\text{Baseline}}{\text{Distance}}
$$
By carefully measuring how much Mars “moved” against distant stars from different observing stations, Cassini could calculate the Earth–Sun distance indirectly.

### **Accuracy**
- Cassini’s estimate got within about 7–10% of the modern value, a major improvement over ancient guesses.


## 3. 18th Century: Transits of Venus

### **Idea**
- Edmond Halley realized that observing Venus crossing the Sun’s disk (a $transit$) from different locations would let astronomers measure the Sun–Earth distance more accurately.
- Multiple worldwide expeditions (1761 and 1769) measured Venus’s exact contact times with the Sun’s edge.

### **Key Formula (in simple terms)**
$$
\text{Parallax angle} = \frac{\text{Baseline on Earth}}{\text{Distance to Venus}}
$$
From this, they worked up to the Sun’s distance, using the geometry of orbits.

### **Accuracy**
- By the late 18th century, the 1 AU value was pinned down to within about 2–3%.
- Challenges included timing errors and weather, but it was still a remarkable leap forward.


## 4. 19th & Early 20th Century Refinements

### **Idea**
- Astronomers used better telescopes, improved star catalogs, and photographic plates to measure $parallax$ more precisely (e.g., of asteroids like Eros).
- Advanced math (orbital mechanics, least-squares) refined the 1 AU measurement further.

### **Accuracy**
- By the early 20th century, estimates got within 1% of the modern value.
- Systematic errors dropped as instruments became more accurate.


## 5. Mid-20th Century to Now: Radar & Spacecraft

### **Idea**
- In the 1960s, scientists bounced radar signals off planets (especially Venus). By timing the signals’ round-trip, they obtained a direct measure of distance.
- Spacecraft tracking (radio signals) then pinpointed the Earth–Sun distance with even greater precision.

### **Accuracy**
- Modern values fix 1 AU at about 149,597,870.7 km.
- The uncertainty is now only in the tens of meters—effectively negligible for most practical purposes.


## Comparative Table of Approximate Accuracy of 1 AU Estimates

| **Era / Method**                   | **Approx. Accuracy vs. Modern Value**     |
|------------------------------------|-------------------------------------------|
| Ancient Greek & Medieval           | Off by factors of 5–10 (or more)          |
| 17th Century (Cassini’s parallax)  | Within ~7–10%                             |
| 18th Century (Transit of Venus)    | Within ~2–3%                              |
| 19th & Early 20th Century          | Better than 1%                            |
| Modern (Radar & Spacecraft)        | Uncertainty of only a few meters (~0.000001%) |


## References (Simplified)
- **Aristarchus & Ptolemy:** Heath, T. (1913). *Aristarchus of Samos*.  
- **Cassini:** Hoskin, M. (1997). *Cambridge Illustrated History of Astronomy*.  
- **Transit of Venus:** Maor, E. (2012). *Venus in Transit*.

**Key takeaway:** Over nearly two thousand years, we’ve gone from very rough guesses to pinpointing the Sun–Earth distance with incredible accuracy—thanks to telescopes, clever geometry, global observations, and finally, radar and spacecraft tracking.

## Q3) Use the period of planets of our solar system to obtain the ratio of distances from the Sun with respect to distance to the Mercury.

## Using Kepler’s Third Law to Find Planetary Distance Ratios (Relative to Mercury)
## IDEA
According to *Kepler’s Third Law*, the square of a planet’s orbital period $T^2$ is proportional to the cube of its average distance from the Sun $a^3$:

$$
T^2 \propto a^3
$$

Hence, in simpler terms: $a \propto T^{\tfrac{2}{3}}$.

If we want each planet’s distance relative to Mercury, we can take the ratio of their orbital periods ($T_\text{planet}$) to Mercury’s orbital period ($T_\text{Mercury}$), then raise that ratio to the $ \tfrac{2}{3} $ power:

$$
\frac{a_\text{planet}}{a_\text{Mercury}}
= \left(\frac{T_\text{planet}}{T_\text{Mercury}}\right)^{\tfrac{2}{3}}.
$$

## 1. Earth

- $T_\text{Mercury} = 0.241 \,\text{years}$
- $T_\text{Earth} = 1.000 \,\text{year}$

1. Calculate the ratio of orbital periods:
   $$
   \frac{T_\text{Earth}}{T_\text{Mercury}}
   = \frac{1.000}{0.241} 
   \approx 4.15.
   $$
2. Raise that to the power of $\tfrac{2}{3}$:
   $$
   \left(\,4.15\,\right)^{\tfrac{2}{3}} 
   \approx 2.59.
   $$

Hence, **Earth is about 2.59 times farther** from the Sun than Mercury.


## 2. Mars

- $T_\text{Mercury} = 0.241 \,\text{years}$
- $T_\text{Mars} = 1.880 \,\text{years}$

1. Calculate the ratio of orbital periods:
   $$
   \frac{T_\text{Mars}}{T_\text{Mercury}}
   = \frac{1.880}{0.241} 
   \approx 7.79.
   $$
2. Raise that to the power of $\tfrac{2}{3}$:
   $$
   \left(\,7.79\,\right)^{\tfrac{2}{3}} 
   \approx 3.95.
   $$

Hence, **Mars is about 3.95 times farther** from the Sun than Mercury.


## 3. Jupiter

- $T_\text{Mercury} = 0.241 \,\text{years}$
- $T_\text{Jupiter} = 11.86 \,\text{years}$

1. Calculate the ratio of orbital periods:
   $$
   \frac{T_\text{Jupiter}}{T_\text{Mercury}}
   = \frac{11.86}{0.241} 
   \approx 49.2.
   $$
2. Raise that to the power of $\tfrac{2}{3}$:
   $$
   \left(\,49.2\,\right)^{\tfrac{2}{3}}
   \approx 13.44.
   $$

Hence, **Jupiter is about 13.44 times farther** from the Sun than Mercury.

## Q4) Find the accuracy of parallax measurement of Hipparcus and GAIA by looking into the papers by the groups. Is it a single number or a distribution? what does the distribution look like?
---
## 1. Hipparcos vs. Gaia: Do They Have a Single Accuracy Number or a Distribution?

When we talk about the accuracy of parallax measurements (the “distance” measurements to stars) from missions like **Hipparcos** and **Gaia**, there is **no single number** that applies to all stars. Instead, the uncertainties follow **distributions** that depend on brightness, crowding of the field, color of the star, and other factors.


## 2. Hipparcos

### **Original Hipparcos Catalog (ESA 1997)**
- Typical parallax uncertainty for bright stars: about $1$ milliarcsecond (mas).
- Fainter stars or tricky cases can have larger uncertainties, sometimes a few mas.

### **New Reduction (van Leeuwen 2007)**
- Refined data processing reduced some systematic errors.
- Typical uncertainties around $0.6$–$1.0$ mas for brighter stars.
- Still varies depending on each star’s observations.

### **What The Uncertainty Distribution Looks Like**
- The errors roughly form a **Gaussian-like “bell curve”** centered on the true value.
- There are “tails” for stars that were harder to measure or had fewer/poorer observations.

## 3. Gaia

### **Gaia DR2 (2018)**
- Bright stars around $G = 15$: typical uncertainty of about $0.03$–$0.04$ mas.
- Fainter stars ($G = 20$): can be around $0.7$ mas or more.
- Notable “zero-point offset” of roughly $0.03$ mas, meaning all measured parallaxes were systematically shifted by that small amount.

### **Gaia EDR3 / DR3**
- Bright stars ($G = 10$–12): as low as $0.02$ mas.
- Fainter stars ($G > 18$–20): can go up to $0.1$–$0.3$ mas or worse.
- Still some small systematic offsets (about $0.02$ mas), but much improved over DR2.

### **Shape of the Distribution**
- Uncertainties get larger as stars get fainter (higher $G$ magnitude).
- Overall, errors typically form a **narrow, near-Gaussian peak**, but again there are wings or tails for special cases (crowded fields, bright/dim limits, etc.).


## 4. Key Takeaways

1. **No Single “Accuracy” Figure:**  
   Each star has its own uncertainty based on how well it was observed.

2. **Distribution of Errors:**  
   Parallax uncertainties look roughly Gaussian for most stars, with tails of outliers.  
   - Hipparcos: $ \sim 0.6$–$1$ mas for best cases, up to a few mas otherwise.  
   - Gaia: $ \sim 0.02$–$0.04$ mas for the brightest, going up to $0.3$–$0.7$ mas for faint stars (depending on data release).

3. **Systematic Offsets:**  
   Gaia data also comes with a “zero-point” offset that must be accounted for in high-precision work.


## 5. References

- **Hipparcos Catalog (1997)**: ESA SP-1200.  
- **Hipparcos New Reduction**: van Leeuwen (2007).  
- **Gaia DR2**: Gaia Collaboration (2018).  
- **Gaia EDR3**: Gaia Collaboration (2020).  
- **Gaia DR3**: Gaia Collaboration (2022).

## Q6) Using sensitivity limit of Hubble Space Telescope, find the maximum distance that can be measured using Red Clump Stars along with the uncertainty.
---
We use the standard **distance modulus** relation:

$$
m - M = 5 \log_{10}\!\biggl(\frac{d}{10 \,\text{pc}}\biggr),
$$

where  
- *m* = apparent magnitude (how bright it looks),  
- *M* = absolute magnitude (intrinsic brightness),  
- *d* = distance in parsecs (pc).

Solving for *d*:

$$
d = 10^{\Bigl(\frac{m - M}{5} + 1\Bigr)} \text{ pc}.
$$


### **1. Typical Values**

- **Absolute Magnitudes of Red Clump Stars**:  
  - $M_{V} \approx +0.6$  
  - $M_{I} \approx -0.28$  
  - $M_{H} \approx -1.4$

- **HST Detection Limits**:  
  - **Visible (V)**: $m \approx 30$  
  - **Infrared (I or H)**: $m \approx 27$

- **Magnitude Uncertainty**: $\pm 0.2\,\text{mag}$


### **2. Distance Calculations**

#### **2.1. Visible Band (V)**

- $M_{V} = +0.6$  
- $m_{V,\text{lim}} = 30$

$$
(m - M) = 30 - 0.6 = 29.4,
$$

$$
5 \log_{10}\!\biggl(\frac{d}{10}\biggr) = 29.4,
$$

$$
\log_{10}\!\biggl(\frac{d}{10}\biggr) = \frac{29.4}{5} = 5.88,
$$

$$
\frac{d}{10} = 10^{5.88} \approx 7.59 \times 10^5,
$$

$$
d \approx 7.59 \times 10^6 \,\text{pc} \approx 7.59\,\text{Mpc}.
$$

#### **2.2. Near Infrared (I)**

- $M_{I} = -0.28$  
- $m_{I,\text{lim}} = 27$

$$
(m - M) = 27 - (-0.28) = 27.28,
$$

$$
5 \log_{10}\!\biggl(\frac{d}{10}\biggr) = 27.28,
$$

$$
\log_{10}\!\biggl(\frac{d}{10}\biggr) = \frac{27.28}{5} = 5.456,
$$

$$
\frac{d}{10} = 10^{5.456} \approx 2.86 \times 10^5,
$$

$$
d \approx 2.86 \times 10^6 \,\text{pc} \approx 2.86\,\text{Mpc}.
$$

#### **2.3. Infrared (H)**

- $M_{H} = -1.4$  
- $m_{H,\text{lim}} = 27$

$$
(m - M) = 27 - (-1.4) = 28.4,
$$

$$
5 \log_{10}\!\biggl(\frac{d}{10}\biggr) = 28.4,
$$

$$
\log_{10}\!\biggl(\frac{d}{10}\biggr) = \frac{28.4}{5} = 5.68,
$$

$$
\frac{d}{10} = 10^{5.68} \approx 4.79 \times 10^5,
$$

$$
d \approx 4.79 \times 10^6 \,\text{pc} \approx 4.79\,\text{Mpc}.
$$


### **3. Uncertainty Estimation**

A $\pm 0.2\,\text{mag}$ uncertainty in $(m - M)$ leads to about a $\pm 10\%$ uncertainty in distance. The reason is:

$$
\Delta (m - M) = \pm 0.2 
\quad \Longrightarrow \quad 
\Delta d \propto 10^{\Bigl(\frac{\pm0.2}{5}\Bigr)} 
= 10^{\pm0.04} 
\approx 1.096 
\approx \pm 10\%.
$$

Thus, our **distance** estimates have roughly $\pm 10\%$ uncertainties.

### **4. Final Summary**

- **Visible (V):**  
  $d \approx 7.59\,\text{Mpc} \,(\pm 10\%)$

- **Near Infrared (I):**  
  $d \approx 2.86\,\text{Mpc} \,(\pm 10\%)$

- **Infrared (H):**  
  $d \approx 4.79\,\text{Mpc} \,(\pm 10\%)$

Because HST can detect fainter objects in the **visible** range, **Red Clump Stars** can be observed the farthest in $V$-band (up to about 7.6 Mpc).

## Q7)Solve numerically the projective motion of a cricket ball (mass = $159\,\text{g}$, diameter = $15\,\text{cm}$) thrown at an angle of $30^\circ$ with velocity $100\,\text{km/h}$. There is an air resistance, $F_D = -\,c\,\tfrac12\,\rho\,A\,v\,v$  
### where,  
### $F_D$ is the drag force,  
### $c \approx 0.5$ is the drag coefficient  
### for a sphere,  
### $\rho$ is the air density, $1.29\,\text{kg/m}^3$.  
### $A$ is the cross sectional area of the projectile.

### Discuss the choice of time step that you can take to get correct numerical answer.

```python
import numpy as np
import matplotlib.pyplot as plt

M = 0.159
D = 0.15
R = D/2
Cd = 0.5
Rho = 1.29
Area = np.pi * R**2
G = 9.81

v0_kmh = 100.0
v0 = v0_kmh / 3.6
th_deg = 30.0
th = np.deg2rad(th_deg)

vx0 = v0 * np.cos(th)
vy0 = v0 * np.sin(th)

def drag(vx, vy):
    s = np.sqrt(vx**2 + vy**2)
    F = 0.5 * Cd * Rho * Area * s**2
    return (-F * vx / s if s > 1e-8 else 0), (-F * vy / s if s > 1e-8 else 0)

dt = 1e-3
t = [0]
x = [0]
y = [0]
vx = [vx0]
vy = [vy0]

while True:
    vx_c, vy_c = vx[-1], vy[-1]
    x_c, y_c = x[-1], y[-1]
    Fx, Fy = drag(vx_c, vy_c)
    ax = Fx / M
    ay = Fy / M - G
    vx_n = vx_c + ax * dt
    vy_n = vy_c + ay * dt
    x_n = x_c + vx_c * dt
    y_n = y_c + vy_c * dt
    t_n = t[-1] + dt
    
    vx.append(vx_n)
    vy.append(vy_n)
    x.append(x_n)
    y.append(y_n)
    t.append(t_n)
    
    if y_n < 0 and t_n > 0:
        break

T_ideal = 2 * v0 * np.sin(th) / G
t_ideal = np.linspace(0, T_ideal, 200)
x_ideal = v0 * np.cos(th) * t_ideal
y_ideal = v0 * np.sin(th) * t_ideal - 0.5 * G * t_ideal**2
plt.figure(figsize = (14, 7))
plt.plot(x, y, label='With Drag')
plt.plot(x_ideal, y_ideal, '--.', label='No Drag')  

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()
plt.grid()
plt.show()

print(f"dt = {dt}, time = {t[-1]:.3f}, range = {x[-1]:.2f}, max height = {max(y):.2f}")
```

So the time scale of $10^{-3}$ was choosen because of following reasons:
1. Typically a normal throw lasts for few seconds in air (considering $ \approx $ 28m/s)
2. To evaluate and showcase the difference between the trajectories a time step of $ \Delta t \approx 0.001$ or $ 0.01 $ is sufficient.
3. Hence a time step less than 1 sec or 0.1 sec will be enough to showcase the effect of drag on the trajectory of the ball.
