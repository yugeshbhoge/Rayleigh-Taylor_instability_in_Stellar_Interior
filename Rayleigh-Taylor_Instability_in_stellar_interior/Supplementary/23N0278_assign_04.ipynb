## Q1. Stefan's Law
The integrated energy density for thermal radiation is,
\begin{equation}
E(T) = \int_{0}^{\infty} E(\nu,T)  d\nu = \Bigl(\frac{8 \pi h}{c^{3}}\Bigr)\int_{0}^{\infty} \nu^{3}  \Bigl(e^{{h \nu}/{kT}}-1\Bigr)^{-1}  d\nu
\end{equation}
Perform this integral using series expansion.

## Solution:

\begin{equation}
E(T) = \int_{0}^{\infty} E(\nu,T)\,d\nu
\;=\; \frac{8\pi h}{c^3}\,\int_{0}^{\infty}\frac{\nu^3}{e^{\frac{h\nu}{k T}}-1}\,d\nu.
\end{equation}

Expand the denominator using exponential series,
\begin{equation}
\frac{1}{e^x - 1} = \sum_{n=1}^{\infty} e^{-n x},
\tag{2}
\end{equation}

with $ x = \frac{h\nu}{k T} $. Hence,
\begin{equation}
\frac{1}{e^{\frac{h\nu}{k T}}-1} = \sum_{n=1}^{\infty} e^{-n\frac{h\nu}{k T}}.
\tag{3}
\end{equation}

Putting this into the integral in eqaution (1):
\begin{equation}
\int_{0}^{\infty}\frac{\nu^3}{e^{\frac{h\nu}{k T}}-1}\,d\nu
= \sum_{n=1}^{\infty}\int_{0}^{\infty}\nu^3\,e^{-n\frac{h\nu}{k T}}\,d\nu.
\tag{4}
\end{equation}

Solving the inner integral using substitution,
$$
x = n\frac{h}{k_B T}\,\nu \quad \Longrightarrow \quad \nu = \frac{x}{n\frac{h}{k T}},\quad d\nu = \frac{dx}{n\frac{h}{k T}}
$$
Then,
$$
\nu^3 = \left(\frac{x}{n\frac{h}{k T}}\right)^3
$$

Thus, the inner integral in equation (4) becomes:
$$
\int_{0}^{\infty}\nu^3\,e^{-n\frac{h\nu}{k T}}\,d\nu
= \frac{1}{\left(n\frac{h}{k T}\right)^4}\int_{0}^{\infty}x^3\,e^{-x}\,dx.
$$

Using the Gamma integral: 
$
\int_{0}^{\infty}x^3\,e^{-x}\,dx = 3! = 6,
$
we get:
$$
\int_{0}^{\infty}\nu^3\,e^{-n\frac{h\nu}{k T}}\,d\nu
= \frac{6}{\left(n\frac{h}{k T}\right)^4}
$$

Plug this back into the sum:
$$
\int_{0}^{\infty}\frac{\nu^3}{e^{\frac{h\nu}{k T}}-1}\,d\nu
= 6\left(\frac{k T}{h}\right)^4 \sum_{n=1}^{\infty}\frac{1}{n^4}.
$$

Riemann zeta function gives:
$
\sum_{n=1}^{\infty}\frac{1}{n^4} = \zeta(4) = \frac{\pi^4}{90}
$

so that
$$
\int_{0}^{\infty}\frac{\nu^3}{e^{\frac{h\nu}{k T}}-1}\,d\nu
= \frac{\pi^4}{15}\left(\frac{k T}{h}\right)^4
$$

Putting this in equation (1), we get:
$$
E(T) = \frac{8\pi h}{c^3}\cdot\frac{\pi^4}{15}\left(\frac{k T}{h}\right)^4
= \frac{8\pi^5 (k T)^4}{15\,c^3\,h^3}
$$

Thus, the final result is:
$$
\boxed{E(T)=\frac{8\pi^5 k^4}{15\,c^3\,h^3}\;T^4.}
$$

## Q2. Classical thermodynamics of radiation gas
Use the expression of energy density ($E(T)=a_{R}T^{4}$) to find the equation of state of a photon gas using first and second laws of thermodynamics


## Solution:
Total internal energy of system can be obatined by multiplying above energy density by volumn,
\begin{equation}
U(T,V)=E(T)V=a_{R}T^4\,V
\end{equation}

First law of thermodynamics:
\begin{equation}
dU=T\,dS-p\,dV
\tag{2}
\end{equation}

Differentiating the equation (1) i.e. $U(T,V)$ w.r.t. $T$ and $V$ gives:
\begin{equation}
dU=\frac{\partial U}{\partial T}dT+\frac{\partial U}{\partial V}dV
=4a_{R}T^3V\,dT+a_{R}T^4\,dV
\tag{3}
\end{equation}

Photon gas undergoes adiabatic process i.e. $ds=0$ gives us,
$$
T^3V=\text{constant},
$$

which implies

$$
\frac{dT}{dV}=-\frac{T}{3V}.
$$

Substitute $dT=-\frac{T}{3V}dV$ in equation (3):
$$
0=4a_{R}T^3V\left(-\frac{T}{3V}dV\right)+a_{R}T^4\,dV
=-\frac{4}{3}a_{R}T^4\,dV+a_{R}T^4\,dV
$$

$$
\left(-\frac{4}{3}+1\right)a_{R}T^4\,dV=-p\,dV,
$$
$$
-\frac{1}{3}a_{R}T^4\,dV=-p\,dV
$$

Cancel the common factor $-dV$ (with $dV\neq 0$) to obtain:

\begin{equation}
p=\frac{1}{3}a_{R}T^4
\tag{4}
\end{equation}

EOS of photon gas is,
\begin{equation}
\boxed{p=\frac{1}{3}E(T)=\frac{1}{3}a_{R}T^4}
\tag{5}
\end{equation}

## Q3. Mean free path for photon
For solar atmosphere, density is $2.1 × 10^{−4} kg m^{−3}$ , opacity coefficient (for 500nm photon) is $0.03 m^{2} kg^{−1}$ . Find the mean free path of photon.

## Solution:
$$
\textbf{Given:} \quad \rho = 2.1 \times 10^{-4}\,\text{kg/m}^3, \quad \kappa = 0.03\,\text{m}^2/\text{kg}.
$$

$$
\quad \lambda = \frac{1}{\rho\,\kappa}
\;\;\Rightarrow\;\;
\quad \lambda = \frac{1}{(2.1 \times 10^{-4})(0.03)}
\;\;\Rightarrow\;\;

(2.1 \times 10^{-4})(0.03) = 6.3 \times 10^{-6}
$$

$$
\lambda: \quad \lambda = \frac{1}{6.3 \times 10^{-6}} \approx 1.59 \times 10^5\,\text{m}
$$

$$
\boxed{\lambda \approx 1.6 \times 10^5\,\text{m} \quad \text{(or about 160 km)}.}
$$


```python

rho = 2.1e-4   # in kg/m^3 
kap = 0.03     # in m^2/kg


mfp = 1 / (rho * kap)

print("Mean free path of a photon: {:.2e} m".format(mfp))

```

## Q 4 Rosseland Mean Opacity of Free-Free Absorption


## Solution
$$
\textbf{Given:}
\quad
\rho\,\kappa_{\nu}
\;=\;
1.32 \times 10^{6}\;\rho^{2}\,g_{\mathrm{ff}}\,\nu^{3}\,T^{-\tfrac12}
\bigl(1 - e^{-\,h\nu/(k_B\,T)}\bigr),
$$

where $\rho$ is the density, $g_{\mathrm{ff}}$ is the free--free Gaunt factor, and $T$ is the temperature.


### 4.1 Deriving $\frac{\partial B_{\nu}}{\partial T}$

Planck function in frequency is given by:
\begin{equation}
B_{\nu}(T)=\frac{2\,h\,\nu^{3}}{c^{2}}\,
\frac{1}{e^{\,h\nu/(k_B\,T)} - 1}.
\tag{1}
\end{equation}

Differentiate w.r.t. $T$:
$$
\frac{\partial B_{\nu}}{\partial T}
=
\frac{2\,h\,\nu^{3}}{c^{2}}
\,
\frac{\partial}{\partial T}
\left[
\frac{1}{e^{\,h\nu/(k_B\,T)} - 1}
\right].
$$

\begin{equation}
\boxed{
    \frac{\partial B_{\nu}}{\partial T}
    =
    \frac{2\,h\,\nu^{3}}{c^{2}}
    \,
    \frac{e^{\,h\nu/(k_B\,T)}}{\bigl(e^{\,h\nu/(k_B\,T)}-1\bigr)^{2}}
    \,
    \frac{h\nu}{k_B\,T^{2}}.
    \tag{2}
}
\end{equation}




```python
import numpy as np
import matplotlib.pyplot as plt

def dBdim(x):
    ex = np.exp(x)
    return x**4 * ex / (ex - 1)**2
xval = np.linspace(1e-3, 20, 500)
yval = dBdim(xval)

# Plot
plt.figure(figsize=(7,5))
plt.plot(xval, yval, 'b-', linewidth=2)
plt.xlabel(r'$x = \frac{h\,\nu}{k_B\,T}$', fontsize=12)
plt.ylabel(r'Dimensionless $\frac{dB_\nu}{dT}$', fontsize=12)
plt.title('Derivative of Planck Function wrt T (Dimensionless)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```


### 4.3 Expressing in Terms of $x$ , where $x = -\frac{h\nu}{k_B T}$.


$$
\nu = -\frac{k_B T}{h}\,x \quad \text{and} \quad d\nu = -\frac{k_B T}{h}\,dx.
$$


Substituting above values of $\nu$ and $d\nu$ in equation (2) we get,
$$
\frac{\partial B_\nu}{\partial T} = \frac{2h\Bigl(-\frac{k_B T}{h}\,x\Bigr)^3}{c^2}\,
\frac{e^{-x}}{(e^{-x}-1)^2}\,
\frac{h\Bigl(-\frac{k_B T}{h}\,x\Bigr)}{k_B T^2}.
$$

After simplifying and using $\frac{e^{-x}}{(e^{-x}-1)^2} = \frac{e^{x}}{(e^{x}-1)^2}$ we get,

$$
\frac{\partial B_\nu}{\partial T} = \frac{2 K_{B}^{3} T^{2}}{h^{2} C^{2}} \frac{x^4\,e^{x}}{(e^{x}-1)^2}.
$$



\begin{equation}
\boxed{
\frac{\partial B_\nu}{\partial T} \propto \frac{x^4\,e^{x}}{(e^{x}-1)^2}.
}
\tag{3}
\end{equation}


### 4.3 Derive an expression for $\frac{\partial B_{\nu}}{\partial T}\,(\rho\,\kappa_{\nu})^{-1}$
$$
h\,\nu \;\sim\; k_{\mathrm{B}}\,T.
$$
Recalling,
$$
\frac{\partial B_{\nu}}{\partial T}
\;=\;
\frac{2\,h\,\nu^{3}}{c^{2}}
\;\frac{e^{\,h\nu/(k_B T)}}{\Bigl(e^{\,h\nu/(k_B T)}-1\Bigr)^{2}}
\;\frac{h\,\nu}{k_B\,T^{2}},
$$

$$
\rho\,\kappa_{\nu}
\;=\;
1.32\times 10^{6}\,\rho^{2}\,g_{\mathrm{ff}}\,\nu^{3}\,T^{-\tfrac12}\,
\Bigl(1 - e^{-\,h\nu/(k_B T)}\Bigr).
$$

Hence,

$$
(\rho\,\kappa_{\nu})^{-1}
\;=\;
\frac{T^{1/2}}{1.32\times10^{6}\,\rho^{3}\,g_{\mathrm{ff}}\,\nu^{3}\,
\bigl(1 - e^{\,x}\bigr)},
\quad
\text{where }
x \;=\;-\frac{h\,\nu}{k_B\,T}.
$$

Multiplying,
$$
\frac{\partial B_{\nu}}{\partial T}
\,(\rho\,\kappa_{\nu})^{-1}
\;=\;
\Bigl[\,
\frac{2\,h\,\nu^{3}}{c^{2}}
\;\frac{e^{x}}{(e^{x}-1)^{2}}
\;\frac{h\,\nu}{k_B\,T^{2}}
\Bigr]
\;\times\;
\Bigl[\,
\frac{T^{1/2}}{1.32\times10^{6}\,\rho^{3}\,g_{\mathrm{ff}}\,
\nu^{3}\,\bigl(1 - e^{x}\bigr)}
\Bigr].
$$

Combine and simplify:

$$
=\;
\frac{2\,h^{2}}{c^{2}\,k_B}
\;\frac{1}{1.32\times10^{6}\,\rho^{3}\,g_{\mathrm{ff}}}
\;\frac{\nu^{4}}{T^{3/2}}
\;\frac{e^{x}}{(1-e^{x})^{2}\,\bigl(1 - e^{x}\bigr)}.
$$


Therefore,
\begin{equation}
\boxed{
\frac{\partial B_{\nu}}{\partial T}
\,(\rho\,\kappa_{\nu})^{-1}
\;\propto\;
\frac{\nu^{4}}{T^{3/2}}
\;\frac{e^{x}}{(1-e^{x})^{3}}
}
\tag{4}
\end{equation}



```python
import numpy as np
import matplotlib.pyplot as mp

def R(x):
    a = x * np.exp(x) / (np.exp(x) - 1)**2
    b = (1 - np.exp(-x)) / x**3
    return a / b

x_a = np.logspace(-2, 1, 200)  
y_a = [R(x) for x in x_a]

mp.figure()
mp.plot(x_a, y_a, label='[∂Bν/∂T]/(ρκν) (dimensionless)')
mp.xlabel(r'$x = \frac{h\nu}{k_B T}$')
mp.ylabel('Dimensionless Ratio')
mp.title('Dimensionless plot of (∂Bν/∂T)/(ρκν) vs. x')
mp.grid(True)
mp.legend()
mp.show()
```

Because the peak in that ratio dominates the Rosseland integral, it demonstrates that the Rosseland mean is determined primarily by the value of $\kappa_{\nu}$ round those frequencies (a few times $\frac{K_{B}T}{h}$) ather than very low or very high frequencies where $R(x)$ is small

### 4.4 Kramers relationship

$$
\frac{\partial B_{\nu}}{\partial T} (\rho\,\kappa_{\nu})^{-1}
\propto \frac{\nu^{4}}{T^{3/2}}\frac{e^{x}}{(1-e^{x})^{3}}

$$

where $x=-\frac{h\nu}{kT}$

The Rosseland mean opacity is defined as
$$
\frac{1}{\kappa}=\frac{\displaystyle \int_{0}^{\infty}\frac{1}{\kappa_{\nu}}\,
\frac{\partial B_{\nu}}{\partial T}\,d\nu}{\displaystyle \int_{0}^{\infty}
\frac{\partial B_{\nu}}{\partial T}\,d\nu}
$$

Since the integrand in the Rosseland mean opacity,
$
\frac{\partial B_{\nu}}{\partial T} (\rho\,\kappa_{\nu})^{-1},
$
peaks at $\nu\sim\frac{kT}{h}$ (or equivalently when $x\sim 2\text{ to }3$), the dominant contribution comes from these frequencies.

For free-free absorption the opacity at frequency $\nu$ is given by
$$
\rho\,\kappa_{\nu}\propto \frac{\rho^2}{\nu^3\,\sqrt{T}}\left(1-e^{-h\nu/(kT)}\right)
$$
At $\nu\sim \frac{kT}{h}$ we have $\nu^3\sim \left(\frac{kT}{h}\right)^3$ and
$
1-e^{-h\nu/(kT)}\sim\mathcal{O}(1),
$
so that
$$
\rho\,\kappa_{\nu}\propto \rho^2\,T^{-3.5}
$$

Dividing by $\rho$ to obtain the opacity per unit mass gives
$$
\boxed{
\kappa\propto \rho\,T^{-3.5}
}
$$

This is the Kramers opacity law for free-free absorption.

## Q5 Eddington Standard Model

## Solution

### 5.1 Luminosity relation


The radiative-diffusion flux equation:

\begin{equation}
F(r) \;=\; \frac{4\,a\,c}{3\,\kappa_0}\,T^{3}\,\frac{dT}{dr} 
\;=\; \frac{c}{3\,\kappa_0}\,\frac{d\bigl(a\,T^{4}\bigr)}{dr}
\tag{1}
\end{equation}


The luminosity is given as:

\begin{equation}
L(r) 
\;=\; 4\,\pi\,r^{2}\,F(r)
\;=\; 4\,\pi\,r^{2}\,\left(\frac{4\,a\,c}{3\,\kappa_0}\,T^{3}\,\frac{dT}{dr}\right)
\;=\; \frac{16\,\pi\,a\,c}{3\,\kappa_0}\;r^{2}\,T^{3}\,\frac{dT}{dr}
\tag{2}
\end{equation}


Converting to Lagrangian (mass) coordinates using hydrostatic equilibrium equations:

\begin{equation}
\frac{dP}{dr} 
\;=\; -\,\rho\,g
\;=\; -\,\frac{G\,M_{r}\,\rho}{r^{2}}.
\tag{3}
\end{equation}
$$
\frac{dP}{dr} 
\;=\; -\,\frac{G\,M_{r}\,\rho}{r^{2}}
\quad\Longrightarrow\quad
\frac{dT}{dr}
\;=\;\frac{dT}{dP}\,\frac{dP}{dr}
\;=\;-\,\frac{G\,M_{r}\,\rho}{r^{2}}\;\frac{dT}{dP}
$$

Putting this back into $eq^{n}$ (2):


$$
L(r) 
\;=\; \frac{16\,\pi\,a\,c}{3\,\kappa_0}\;r^{2}\,T^{3}\,\left(\frac{dT}{dr}\right)
\;=\; \frac{16\,\pi\,a\,c}{3\,\kappa_0}\;r^{2}\,T^{3}
\left(-\,\frac{G\,M_{r}\,\rho}{r^{2}}\;\frac{dT}{dP}\right)
\;=\;-\;\frac{16\,\pi\,a\,c\,G}{3\,\kappa_0}\;M_{r}\,\rho\;T^{3}\;\frac{dT}{dP}
$$

No we can write $\frac{dT}{dP}$ as $\frac{dT}{dP}\;=\;\frac{T}{P}\;\frac{d\ln T}{d\ln P}$



$$
L(r) 
\;=\;
-\;\frac{16\,\pi\,a\,c\,G}{3\,\kappa_0}\;M_{r}\,\rho\;\frac{T^{4}}{P}\;\frac{d\ln T}{d\ln P}.
$$

In the Eddington standard model (dropping the minus sign for outward luminosity), we get,

\begin{equation}
\boxed{
L(r) \;=\;
\frac{16\,\pi\,a\,c\,G\,T^{4}}{3\,P\kappa_0}\;M_{r}\;\frac{d\ln T}{d\ln P}.
}
\tag{3}
\end{equation}


### 5.2 Including photon EOS


$ \textbf{Given:} \quad P_{\mathrm{rad}} \;=\; \frac{a\,T^{4}}{3}$


Radiation momentum equation is,


\begin{equation}
\frac{dP_{\mathrm{rad}}}{dr} \;=\; -\,\frac{\kappa\,\rho}{c}\,F(r)
\;=\; -\,\frac{\kappa\,\rho}{c}\,\frac{L(r)}{4\,\pi\,r^{2}}.
\tag{4}
\end{equation}

The hydrostatic equilibrium equation is,


\begin{equation}
\frac{dP}{dr} 
\;=\; -\,\rho\,g
\;=\; -\,\frac{G\,M_{r}\,\rho}{r^{2}}.
\tag{3}
\end{equation}

Divide equation (4) by equation (3) 

$$
\frac{dP_{\mathrm{rad}}}{dP}
\;=\;
\frac{\displaystyle \frac{dP_{\mathrm{rad}}}{dr}}{\displaystyle \frac{dP}{dr}}
\;=\;
\frac{-\,\frac{\kappa\,\rho}{c}\,\frac{L(r)}{4\,\pi\,r^{2}}}{-\,\frac{G\,M_{r}\,\rho}{r^{2}}}
\;=\;
\frac{\kappa\,L(r)}{4\,\pi\,c\,G\,M_{r}}.
$$


Relating $\frac{L(r)}{M_{r}}$ to total $\frac{L}{M}$ we get, $\frac{L(r)}{M_{r}}\;=\;\left(\frac{L(r)}{L}\right)\left(\frac{M}{M_{r}}\right)\left(\frac{L}{M}\right)$

Therefore we get,

\begin{equation}
\boxed{
\frac{dP_{\mathrm{rad}}}{dP}
\;=\;
\frac{\kappa}{4\,\pi\,c\,G}
\;\frac{L(r)}{M_{r}}
\;=\;
\frac{\kappa}{4\,\pi\,c\,G}
\left(\frac{L(r)}{L}\,\frac{M}{M_{r}}\,\frac{L}{M}\right)
\;=\;
\frac{L\,\kappa}{4\,\pi\,c\,G\,M}
\;\frac{\frac{L(r)}{L}}{\frac{M_{r}}{M}}.
}
\tag{5}
\end{equation}



### 5.3 

Given notation: $\quad \eta(r) \;=\; \frac{\frac{L(r)}{L}}{\frac{M_r}{M}}\quad\Longrightarrow\quad\eta(r) \;=\; \frac{L(r)/L}{M_r/M}$

From subpart (2), we have: 
$$
\quad
\frac{dP_{\mathrm{rad}}}{dP}
\;=\;
\frac{L\,\kappa}{4\,\pi\,c\,G\,M}\,\eta(r)
$$

We integrate this expression from $P=0\text{ (surface) to }P(r)\text{ (at radius }r\text{)}$

$$
P_{\mathrm{rad}}(r) \;-\; P_{\mathrm{rad}}(0)
\;=\;
\int_{0}^{P(r)} \frac{dP_{\mathrm{rad}}}{dP}\,dP
\;=\;
\int_{0}^{P(r)} 
\frac{L\,\kappa}{4\,\pi\,c\,G\,M}\,\eta(r')
\;dP'
$$

Assuming $P_{\mathrm{rad}}(0)=0$ at the surface, we have,


$$
P_{\mathrm{rad}}(r)
\;=\;
\frac{L}{4\,\pi\,c\,G\,M}
\;\int_{0}^{P(r)} \kappa\,\eta(r')\;dP'
$$

Define an appropriate average of $\kappa\,\eta(r')$ over the interval $[0,\,P(r)]$


$$
\bigl\langle \kappa\,\eta \bigr\rangle_r
\;\equiv\;
\frac{1}{P(r)}\,\int_{0}^{P(r)} \kappa\,\eta(r')\;dP'
$$

Then we can write:


$$
P_{\mathrm{rad}}(r)
\;=\;
\frac{L}{4\,\pi\,c\,G\,M}\;\bigl\langle \kappa\,\eta \bigr\rangle_r\;P(r)
$$

The effective mean opacity  $\kappa(r)
\;\equiv\;
\frac{1}{P(r)}\,\int_{0}^{P(r)} \kappa\;dP'
\quad\text{and similarly assumes}\quad
\eta(r')\approx \eta(r)
$


Under suitable assumptions (e.g., linear relations in $r$ or constant $\eta$), one obtains:


$$
\boxed{
P_{\mathrm{rad}}(r)
\;=\;
\frac{L}{4\,\pi\,c\,G\,M}
\;\bigl[\kappa(r)\,r\bigr]\;P(r)
}
$$


where $\kappa(r)$ is as an average over the region $0\le P'\le P(r)$, and $r$ is the radial coordinate.
